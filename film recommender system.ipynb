```python
def convert(gen):
    #get the "real" genres helper function
    lis=[]
    for i in ast.literal_eval(gen):
        lis.append(i['name'])
    return lis
```

```python
def con4(gen):
    #get the "real" genres helper function
    lis=[]
    count=0
    for i in ast.literal_eval(gen):
        if count!=4:
            count+=1
            lis.append(i['name'])
        else:
            break
    return lis
```

```python
movies['cast']=movies['cast'].apply(con4)
```

```python
movies.head()
```

```python
def essential(gen):
    #get essential roles in film crew helper function
    lis=[]
   
    for i in ast.literal_eval(gen):
        if i['job']=='Director' or i['job']=='Original Music Composer':
            lis.append(i['name'])
            
    return lis
```

```python
movies['crew'].apply(essential)
```

```python
movies['crew']= movies['crew'].apply(essential)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#eliminate space
movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
#define new dataframe
new_df=movies[['id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
#helper function: to get the stem word. ex. loves->love 
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
v=cv.fit_transform(new_df['tags']).toarray()
```

```python
v
```

```python
v[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(v)
```

```python
#main function
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0] #id
    dist=similarity[movie_index]
    rec_movies=sorted(list(enumerate(dist)),reverse=True,key=lambda x:x[1])[1:7]
    for i in rec_movies:
        print(new_df.iloc[i[0]].title)
    
```

```python
sorted(similarity[0],reverse=True)
```

```python
recommend('Batman')
```
